```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
df = pd.read_csv('2255236.csv')
```

```python
#Checking out the data
df.head()
```

```python
#Getting rid of irrelevant columns.
df = df.drop(['STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'TSUN', 'WDF2', 'WDF5'], axis=1)
```

```python
df.columns
```

```python
#Fixing colmn names so they make sense.
df.columns = ['Date', 'Average Daily Wind Speed Mph', 'Fastest mile Wind Speed time', 'Peak gust time', 'Percipitation', 'Snowfall', 'Snow depth', 'Average Temp', 'Max Temp', 'Min Temp', 'Fastest 2 min wind', 'Fastest 5 min wind', 'Fog/Ice Fog', 'Heavy Fog','Thunder', 'Hail', 'Smoke/Haze', 'Tornado']
```

```python
df.head()
```

```python
df.describe()
```

```python
#Some na values changed to zero for mostly bool columns.
df[['Snowfall', 'Snow depth', 'Fog/Ice Fog', 'Heavy Fog','Thunder', 'Hail', 'Smoke/Haze', 'Tornado']] = df[['Snowfall', 'Snow depth', 'Fog/Ice Fog', 'Heavy Fog','Thunder', 'Hail', 'Smoke/Haze', 'Tornado']].fillna(0)
```

```python
#ffill on the rest.
df[['Average Daily Wind Speed Mph', 'Fastest mile Wind Speed time', 'Peak gust time', 'Percipitation', 'Fastest 2 min wind', 'Fastest 5 min wind', 'Max Temp', 'Min Temp']] = df[['Average Daily Wind Speed Mph', 'Fastest mile Wind Speed time', 'Peak gust time', 'Percipitation', 'Fastest 2 min wind', 'Fastest 5 min wind', 'Max Temp', 'Min Temp']].ffill() 
```

```python
df['Date'].max()
```

```python
df.info()
```

```python
#Add the max and min for the day and divide by 2 to get the avg for any days its missing.
df['Average Temp'] = df.apply(
    lambda row: (row['Max Temp']+row['Min Temp'])/2 if np.isnan(row['Average Temp']) else row['Average Temp'],
    axis=1
)
```

```python
df.info()
```

```python
df['Fastest mile Wind Speed time'] = df['Fastest mile Wind Speed time'].astype(int)
df['Peak gust time'] = df['Peak gust time'].astype(int)
```

```python
#Made these bools for convenience sake
df['Fog/Ice Fog'] = df['Fog/Ice Fog'].astype('bool')
df['Heavy Fog'] = df['Heavy Fog'].astype('bool')
df['Thunder'] = df['Thunder'].astype('bool')
df['Hail'] = df['Hail'].astype('bool')
df['Smoke/Haze'] = df['Smoke/Haze'].astype('bool')
df['Tornado'] = df['Tornado'].astype('bool')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```
